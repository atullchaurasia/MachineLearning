```python
#importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading dataset
from sklearn.datasets import fetch_california_housing
california = fetch_california_housing()
```

```python
#getting attributes of california
dir(california)
```

```python
#Describing dataframe from the data
df=pd.DataFrame(california.data,columns=california.feature_names)
```

```python
df.head(5)
```

```python
# adding a new column 'target' from california.target
df['target']=california.target
```

```python
df.head(2)
```

```python
df.info()
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
print(X_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model_pred =model.predict(X_test)
```

```python
mse = np.mean((model_pred - y_test)**2)
mse
```

```python
# Putting together the coefficient and their columns
model_coeff = pd.DataFrame()
model_coeff['Columns'] = df.columns
model_coeff['Coefficient Values'] = pd.Series(model.coef_)
model_coeff
```

```python
from sklearn.linear_model import Ridge
ridge_model = Ridge(alpha = 1)
ridge_model.fit(X_train, y_train)
```

```python
y_pred = ridge_model.predict(X_test)
```

```python
# Putting together the coefficient and their columns
ridge_coeff=pd.DataFrame()
ridge_coeff['columns']=df.columns
ridge_coeff['Coefficient estimates']=pd.Series(ridge_model.coef_)
print(ridge_coeff)
```

```python
from sklearn.linear_model import Lasso
lasso_model = Lasso(alpha = 1)
lasso_model.fit(X_train, y_train)
```

```python
y_pred = lasso_model.predict(X_test)
```

```python
lasso_mse = np.mean((y_pred - y_test)**2)
lasso_mse
```

```python
# Putting together the coefficient and their columns
lasso_coeff=pd.DataFrame()
lasso_coeff['columns']=df.columns
lasso_coeff['Coefficient estimates']=pd.Series(lasso_model.coef_)
print(lasso_coeff)
```

```python
type(lasso_coeff)
```

```python
from sklearn.linear_model import ElasticNet
elastic = ElasticNet(alpha= 1)
elastic.fit(X_train, y_train)
```

```python
y_pred2 = elastic.predict(X_test)
```

```python
elastic_mse = np.mean((y_pred2 - y_test)**2)
elastic_mse
```

```python
elastic_coeff=pd.DataFrame()
elastic_coeff['columns']=df.columns
elastic_coeff['Coefficient estimates']=pd.Series(elastic.coef_)
print(elastic_coeff)
```

```python
type(elastic_coeff)
```

```python
print(ridge_coeff)
print(lasso_coeff)
print(elastic_coeff)
```
